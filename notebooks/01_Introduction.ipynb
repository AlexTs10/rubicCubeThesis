## Part 1: The Rubik's Cube

### What is the Rubik's Cube?

The Rubik's Cube is a 3D combination puzzle invented by ErnÅ‘ Rubik in 1974. It has:
- **6 faces** (Up, Down, Front, Back, Left, Right)
- **54 colored facelets** (9 per face)
- **26 movable pieces** (8 corners, 12 edges, 6 centers)

### The Numbers

- **Total states**: 43,252,003,274,489,856,000 (43 quintillion!)
- **God's Number**: 20 (maximum moves to solve any scramble)
- **Valid permutations**: Only 1/12 of all possible arrangements are reachable

### Singmaster Notation

Moves are denoted by face letters:
- **U** (Up), **D** (Down), **F** (Front), **B** (Back), **L** (Left), **R** (Right)
- **'** suffix = counter-clockwise (e.g., U')
- **2** suffix = 180Â° turn (e.g., U2)

## Part 2: Creating a Cube

```python
# Create a solved cube
cube = RubikCube()

# Check if solved
print(f"Is cube solved? {cube.is_solved()}")

# Display cube state (2D unfolded net)
print("\nCube state:")
print(display_cube_unfolded(cube))
```

```python
# 3D visualization
fig = visualize_cube_3d(cube, view_angles=(30, 45))
fig.suptitle("Solved Rubik's Cube", fontsize=14)
plt.show()
```

## Part 3: Applying Moves

```python
# Apply some moves
moves = ["R", "U", "R'", "U'"]

for move in moves:
    cube.apply_move(move)
    print(f"Applied {move}")

print(f"\nIs cube solved? {cube.is_solved()}")

# Visualize scrambled cube
fig = visualize_cube_3d(cube, view_angles=(30, 45))
fig.suptitle("After R U R' U'", fontsize=14)
plt.show()
```

## Part 4: Scrambling

```python
# Create a fresh cube and scramble it
cube = RubikCube()
cube.scramble(n_moves=10, seed=42)

# Get the scramble sequence
scramble_moves = getattr(cube, '_scramble_moves', [])
print(f"Scramble sequence: {' '.join(scramble_moves)}")
print(f"Scramble depth: {len(scramble_moves)} moves")

# Visualize
fig = visualize_cube_3d(cube, view_angles=(30, 45))
fig.suptitle("Scrambled Cube (10 moves)", fontsize=14)
plt.show()
```

## Part 5: The Three Algorithms

### Thistlethwaite's Algorithm (1981)

**Approach:** Group theory - reduces cube through 4 nested subgroups  
**Speed:** âš¡ Very fast (0.2-0.5s)  
**Solution:** 30-52 moves (sub-optimal)  
**Memory:** ðŸ’¾ Low (~2 MB)

### Kociemba's Algorithm (1992)

### Korf's IDA* (1997)

## Part 6: Quick Solve Example

```python
from src.thistlethwaite import ThistlethwaiteSolver
import time

# Create a scrambled cube
cube = RubikCube()
cube.scramble(n_moves=10, seed=42)

print("Solving with Thistlethwaite...")

# Solve
solver = ThistlethwaiteSolver(use_pattern_databases=False)
start = time.time()
solution = solver.solve(cube.copy())
elapsed = time.time() - start

print(f"\nâœ“ Solution found!")
print(f"  Moves: {len(solution)}")
print(f"  Time: {elapsed:.3f}s")
print(f"\nSolution sequence:")
print(' '.join(solution))

# Verify solution
test_cube = cube.copy()
for move in solution:
    test_cube.apply_move(move)

print(f"\nVerification: Cube is {'SOLVED âœ“' if test_cube.is_solved() else 'NOT SOLVED âœ—'}")
```
