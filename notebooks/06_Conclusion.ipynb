## 3. Visual Comparison

```python
# Create comprehensive comparison data
algorithms = ['Thistlethwaite', 'Kociemba', 'Korf IDA*']

# Typical metrics (based on benchmarks)
avg_moves = [40, 17, 18.5]
avg_time_ms = [500, 2000, 8000]  # milliseconds for better scale
memory_mb = [2, 80, 45]

# Create figure with 3 subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16, 5))
colors = ['#3498db', '#2ecc71', '#9b59b6']

# 1. Solution Length
bars1 = ax1.bar(algorithms, avg_moves, color=colors, alpha=0.7, edgecolor='black', linewidth=1.5)
ax1.axhline(y=20, color='red', linestyle='--', linewidth=2, label="God's Number: 20")
ax1.set_ylabel('Average Moves', fontsize=13, fontweight='bold')
ax1.set_title('Solution Length Comparison', fontsize=15, fontweight='bold')
ax1.legend(fontsize=11)
ax1.grid(axis='y', alpha=0.3)
for i, (bar, val) in enumerate(zip(bars1, avg_moves)):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height,
             f'{val:.1f}', ha='center', va='bottom', fontsize=12, fontweight='bold')

# 2. Speed (log scale)
bars2 = ax2.bar(algorithms, avg_time_ms, color=colors, alpha=0.7, edgecolor='black', linewidth=1.5)
ax2.set_ylabel('Average Time (ms, log scale)', fontsize=13, fontweight='bold')
ax2.set_title('Execution Speed Comparison', fontsize=15, fontweight='bold')
ax2.set_yscale('log')
ax2.grid(axis='y', alpha=0.3, which='both')
for bar, val in zip(bars2, avg_time_ms):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height,
             f'{val:.0f}ms', ha='center', va='bottom', fontsize=10)

# 3. Memory Usage
bars3 = ax3.bar(algorithms, memory_mb, color=colors, alpha=0.7, edgecolor='black', linewidth=1.5)
ax3.set_ylabel('Memory (MB)', fontsize=13, fontweight='bold')
ax3.set_title('Memory Usage Comparison', fontsize=15, fontweight='bold')
ax3.grid(axis='y', alpha=0.3)
for bar, val in zip(bars3, memory_mb):
    height = bar.get_height()
    ax3.text(bar.get_x() + bar.get_width()/2., height,
             f'{val} MB', ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()

print("\nðŸ† Winner by Category:")
print("  â€¢ Fewest moves: Kociemba (17 avg)")
print("  â€¢ Fastest: Thistlethwaite (500ms avg)")
print("  â€¢ Least memory: Thistlethwaite (2 MB)")
print("  â€¢ Most optimal: Korf (guaranteed â‰¤20)")
print("  â€¢ Best balance: Kociemba â­")
```

## 4. Trade-off Analysis

### The Speed-Quality Triangle

There are three competing factors:
1. **Speed** - How fast is the solution?
2. **Quality** - How short is the solution?
3. **Memory** - How much RAM is needed?

**You can optimize 2 out of 3:**

- **Thistlethwaite:** Speed + Low Memory â†’ Poor Quality
- **Kociemba:** Speed + Quality â†’ High Memory
- **Korf:** Quality + Optimal â†’ Slow Speed

### Use Case Matrix

| Scenario | Best Algorithm | Why |
|----------|----------------|-----|
| **Speed cubing competition** | Kociemba | Near-optimal + fast |
| **Educational demo** | Thistlethwaite | Fast + easy to understand |

## 5. Multi-Dimensional Comparison

```python
# Create radar/spider chart for multi-dimensional comparison
categories = ['Speed', 'Solution Quality', 'Memory Efficiency', 
              'Ease of Implementation', 'Educational Value', 'Practical Use']
N = len(categories)

# Scores (0-10 scale)
thistlethwaite_scores = [9, 3, 9, 6, 10, 5]
kociemba_scores = [7, 9, 3, 4, 8, 10]
korf_scores = [3, 10, 7, 3, 8, 4]

# Compute angle for each axis
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]  # Complete the circle

# Complete the data circles
thistlethwaite_scores += thistlethwaite_scores[:1]
kociemba_scores += kociemba_scores[:1]
korf_scores += korf_scores[:1]

# Create plot
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(projection='polar'))

# Plot data
ax.plot(angles, thistlethwaite_scores, 'o-', linewidth=2, label='Thistlethwaite', color='#3498db')
ax.fill(angles, thistlethwaite_scores, alpha=0.15, color='#3498db')

ax.plot(angles, kociemba_scores, 'o-', linewidth=2, label='Kociemba', color='#2ecc71')
ax.fill(angles, kociemba_scores, alpha=0.15, color='#2ecc71')

ax.plot(angles, korf_scores, 'o-', linewidth=2, label='Korf', color='#9b59b6')
ax.fill(angles, korf_scores, alpha=0.15, color='#9b59b6')

# Set category labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, size=12)

# Set y-axis
ax.set_ylim(0, 10)
ax.set_yticks([2, 4, 6, 8, 10])
ax.set_yticklabels(['2', '4', '6', '8', '10'], size=10)
ax.grid(True)

# Add legend and title
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1), fontsize=12)
plt.title('Multi-Dimensional Algorithm Comparison', size=16, fontweight='bold', pad=20)

plt.tight_layout()
plt.show()

print("\nðŸ“Š Interpretation:")
print("  â€¢ Larger area = Better overall performance")
print("  â€¢ Thistlethwaite: Strong in speed and education")
print("  â€¢ Kociemba: Most balanced (largest area)")
print("  â€¢ Korf: Excels in solution quality only")
```

## 6. Lessons Learned

### Technical Lessons

1. **Group theory is powerful**
   - Thistlethwaite showed how mathematical structure enables efficient solving
   - Subgroup chains reduce problem complexity

2. **Heuristics make or break search**
   - Pattern databases dramatically improve IDA* performance
   - Admissible heuristics guarantee optimality

3. **Optimality is expensive**
   - Moving from near-optimal to optimal costs 10-100x time
   - Often not worth it for practical applications

4. **Memory-time trade-offs are fundamental**
   - More memory â†’ better heuristics â†’ faster search
   - Must choose based on constraints

5. **Simplicity has value**
   - Kociemba's 2 phases beat Thistlethwaite's 4
   - Sometimes fewer phases with better search wins

### Implementation Lessons

1. **Testing is critical**
   - Cube algorithms have many edge cases
   - Comprehensive test suites essential

2. **Coordinate systems matter**
   - Efficient encodings speed up everything
   - Symmetry reductions help

3. **Benchmarking requires care**
   - Need diverse test cases
   - Statistical significance important

4. **Visualization aids understanding**
   - 3D cube views helpful for debugging
   - Move animations show algorithm behavior

### Project Management Lessons

1. **Start simple, iterate**
   - Implemented Thistlethwaite first
   - Built on experience for complex algorithms

2. **Documentation saves time**
   - Clear API docs essential
   - Examples help future users

3. **Modular design pays off**
   - Shared cube representation
   - Reusable components

4. **Performance testing early**
   - Found bottlenecks during development
   - Optimized hot paths
